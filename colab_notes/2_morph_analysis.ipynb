## 形態素解析

### janome

```python
!pip install janome
```

```python
from janome.tokenizer import Tokenizer

# 分かち書き
first_text = 'この物語の主人公アレクセイ・フョードロヴィッチ・カラマゾフの伝記にとりかかるに当たって、自分は一種の懐疑に陥っている。'
tokenizer = Tokenizer(wakati=True)
token_ls = list(tokenizer.tokenize(first_text))
print(token_ls)
```

```python
# 品詞情報も表示
tokenizer = Tokenizer(wakati=False)  # defaultでFalseだが，あえて明示的に書く
for t in tokenizer.tokenize(first_text):
    print(t)
```

```python
# 形態素の表層系と品詞のみ表示
tokenizer = Tokenizer(wakati=False)  # defaultでFalseだが，あえて明示的に書く
for t in tokenizer.tokenize(first_text):
    word_class = t.part_of_speech.split(',')[0]
    print(f'{t.surface}\t{word_class}')
```

```python
import ginza

ginza.set_split_mode(nlp, 'A')
doc = nlp('自然言語処理')
token_ls = list(doc)
print('mode A: ', token_ls)

ginza.set_split_mode(nlp, 'C')
doc = nlp('自然言語処理')
token_ls = list(doc)
print('mode C: ', token_ls)
```

```python
doc = nlp(first_text)
token_ls = list(doc)

for t in token_ls:
     print(t.text, t.pos_, t.tag_, sep='\t')
```
